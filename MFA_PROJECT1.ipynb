Mean Imputation: Fill missing values with the column mean

```python
df_mean_imputed = df.fillna(df.select_dtypes(include='number').mean())


print("\nDataFrame after mean imputation:")
print(df_mean_imputed.head())

#  saving  the imputed DataFrame back to the original file
df_mean_imputed.to_csv('imputed_file.csv', index=False)
```

median

```python
df_median_imputed = df.fillna(df.select_dtypes(include='number').median())
print("\nDataFrame after median imputation:")
print(df_median_imputed.head())
```

Mode

```python
df_mode_imputed = df.copy()
for col in df.columns:
    df_mode_imputed[col] = df_mode_imputed[col].fillna(df[col].mode()[0])
print("\nDataFrame after mode imputation:")
print(df_mode_imputed.head())
```

# 9. **Handling Imbalanced Data**

Checking Class Distribution

```python
print(df['Electric Vehicle Type'].value_counts())

```

(i) Oversampling the Minority Class

```python
X = df.drop(columns=['Electric Vehicle Type'])
y = df['Electric Vehicle Type']
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
print("Class distribution after oversampling:", Counter(y_resampled))
```

(ii) Under-sampling the Majority Class

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)
print(pd.Series(y_resampled).value_counts())
```

( ii ) Normality Testing (Using the Shapiro-Wilk Test)

```python
from scipy.stats import shapiro
for column in numerical_features:
    stat, p = shapiro(df[column].dropna())
    print(f"Shapiro-Wilk Test for {column}: Statistics={stat:.3f}, p-value={p:.3f}")
    if p > 0.05:
        print(f"{column} is normally distributed.")
    else:
        print(f"{column} is not normally distributed.")
```

 ( iii ) Homoscedasticity (Using Levene's Test)

```python
print(df.columns)
```

```python
from scipy.stats import levene
group1 = df[df['County'] == 'Group1']['Range_Category']
group2 = df[df['Base MSRP'] == 'Group2']['DOL Vehicle ID']
stat, p = levene(group1, group2)
print(f"Levene's Test: Statistics={stat:.3f}, p-value={p:.3f}")
if p > 0.05:
    print("Variances are equal across groups (Homoscedasticity).")
else:
    print("Variances are not equal across groups (Heteroscedasticity).")
```
